# **Importando bibliotecas**

```python
# Importando bibliotecas necessárias
from pydataset import data
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

# **Carregando os dados do banco**

```python
# Carregando o conjunto de dados do banco de Duncan
duncan = data('Duncan')
```

# **Selecionando as variáveis**

```python
# Selecionando apenas variáveis quantitativas para a análise
quantitative_variables = ['income', 'education', 'prestige']
```

# **Matriz de correlação**

```python
# Calculando e exibindo a matriz de correlação
correlation_matrix = duncan[quantitative_variables].corr()
print("Matriz de Correlação:")
print(correlation_matrix)
```

# **Construindo o gráfico de dispersão e a reta**

## **Relação entre income e education**

### **Imprimindo o gráfico e a reta**

```python
# Criando o gráfico de dispersão
sns.scatterplot(x=quantitative_variables[0], y=quantitative_variables[1], data=duncan)

# Ajustando uma regressão linear
model = LinearRegression()
X = duncan[quantitative_variables[0]].values.reshape(-1, 1)
y = duncan[quantitative_variables[1]].values
model.fit(X, y)

# Plotando a reta de regressão
plt.plot(X, model.predict(X), color='red')

# Calculando o coeficiente de determinação
r2 = r2_score(y, model.predict(X))

 # Exibindo a expressão da reta de regressão estimada e o coeficiente de determinação
equation = f"{quantitative_variables[1]} = {model.coef_[0]:.2f}*{quantitative_variables[0]} + {model.intercept_:.2f}"
plt.title(f'Reta de Regressão: {equation}, R² = {r2:.2f}')

 # Exibindo o gráfico
plt.show()
```
